```python
import numpy as np
from scipy.integrate import quad
import matplotlib.pyplot as plt
```

I have taken 
$f(y) = y$ \
$a = 2$ \
$b=5$ \
$n=100$

```python

def integrand(y, i, b):
    return (y*np.sin(i*np.pi*y/b))
```

```python
def func(a,b,n):

    x = np.linspace(0,a,n)
    y = np.linspace(0,b,n)
    X,Y = np.meshgrid(x,y)
    
    U = 0
    for i in range(1,n):
  
        I = quad(integrand, 0, b, args=(i,b))
        c = i*np.pi*a/b
        d = b*np.sinh(c)
        r = 2/d
        q = I[0]*r
        u = q * np.sinh(i*np.pi*X/b) * np.sin(i*np.pi*Y/b)
        U = u + U
    return(U)
```

```python
U = func(2,5,100)
```

```python
x = np.linspace(0,a,n)
y = np.linspace(0,b,n)
X,Y = np.meshgrid(x,y)
```

```python
fig = plt.figure()
plt.contour(X,Y,U, cmap='plasma')
plt.colorbar()
plt.show()
```

```python
fig = plt.figure()
plt.contourf(X,Y,U, cmap='viridis')
plt.colorbar()
#plt.axis('scaled')
plt.savefig('Contour.png')
plt.show()
```

```python
ax = plt.axes(projection ='3d')
ax.plot_surface(X,Y,U, cmap = 'viridis')
ax.view_init(45,45)
ax.set_xlabel('x')
ax.set_zlabel('u')
ax.set_ylabel('y')
plt.tight_layout()
plt.show()
```

```python
ax = plt.axes(projection ='3d')
ax.plot_surface(X,Y,U, cmap = 'viridis')

ax.set_xlabel('x')
ax.set_zlabel('u')
ax.set_ylabel('y')
ax.view_init(20,120)
plt.tight_layout()
plt.show()
```

```python
#Wired View

ax = plt.axes(projection='3d')
ax.contour3D(X,Y,U,10, cmap='plasma')
ax.view_init(45,45)
plt.show
```
